```python
%%writefile reducer2.py
#!/usr/bin/env python
#Answer for reducer2.py
import sys
from collections import defaultdict

author_comment_input_pairs = sys.stdin.readlines()
author_comment_accumulator = defaultdict(lambda: [])

for row in author_comment_input_pairs:
    key_value_pair = row.split('\t',1)
    if len(key_value_pair) != 2:
        continue
        
    author = key_value_pair[0]
    comment = key_value_pair[1]
    author_comment_accumulator[author].append(comment)

for (key, value) in author_comment_accumulator.items():
    print(key + "\t" + str(value))
```

```python
%%writefile reducer2.py
#!/usr/bin/env python2.7
#Answer for reducer.py

###the newest version(write to a function)
import sys
from collections import defaultdict
import json
import os

# test list
comments_input_pairs=[
    'John	a',
    'Mary	b',
    'John	c',
    'Mary	d',
    'Jane	e']


def get_all_authors_comments(comments_input_pairs):
    comments_accumulator = defaultdict(lambda: '')
    for row in comments_input_pairs:
        comments_key_value_pair = row.split("\t", 1)  #######row is string, split every row into input_pairs into 2(second parameter 1+1) parts

        if len(comments_key_value_pair) != 2:
            continue

        author = comments_key_value_pair[0]
        comment= comments_key_value_pair[1]

        if(comments_accumulator[author]==''):
            comments_accumulator[author] = comments_accumulator[author] +comment
        else:
            comments_accumulator[author] = comments_accumulator[author] +','+comment

    return comments_accumulator
# for (key, value) in  datetime_accumulator.items():
#         print(key + "\t" + str(value))

def get_comments_one_author(author,comments_input_pairs):
    comments_accumulator=get_all_authors_comments(comments_input_pairs)
    comments=comments_accumulator[author]
    comment_author_dict={author:comments}
    comment_json_array = json.dumps(comment_author_dict,sort_keys = True, indent = 2)
    return comment_json_array

comments_input_pairs = sys.stdin.readlines()
author=str(os.environ.get('author'))
print('author='+author)
# comment_json_array=get_comments_one_author('John',comments_input_pairs)
# comment_json_array=get_comments_one_author("Leonel Hernandez",comments_input_pairs)
comment_json_array=get_comments_one_author(author,comments_input_pairs)
# print(len(comment_json_array))
print(comment_json_array)
```
